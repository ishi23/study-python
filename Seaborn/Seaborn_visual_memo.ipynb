# ***Seabornで見た目をよくする***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## set_theme

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
f, (a1, a2) = plt.subplots(1, 2, figsize=(10,5))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', ax=a1)
sns.scatterplot(data=iris, x='petal_length', y='petal_width'\
                , hue='species', style='species', size='sepal_length', ax=a2)
```

```python
f, (a1, a2) = plt.subplots(1, 2, figsize=(10,5))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', ax=a1)
sns.scatterplot(data=iris, x='petal_length', y='petal_width'\
                , hue='species', style='species', size='sepal_length', ax=a2)
```

```python
a2.legend()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
f, (a1, a2) = plt.subplots(1, 2, figsize=(10, 5))
sns.boxenplot(data=titanic, x='pclass', y='age', ax=a1)
sns.boxenplot(data=titanic, x='pclass', y='age', hue='sex', ax=a2)
```

```python
sns.catplot(data=titanic, x='pclass', y='age', kind='boxen', hue='survived', col='sex')
```

```python
f, (a1, a2) = plt.subplots(1, 2, figsize=(10, 5))
sns.boxenplot(data=titanic, x='pclass', y='age', ax=a1)
sns.boxenplot(data=titanic, x='pclass', y='age', hue='sex', ax=a2)

a2.legend(loc=1, bbox_to_anchor=(1.3, 0.5))
```

```python
f, (a1, a2) = plt.subplots(1, 2, figsize=(10, 5))
sns.boxenplot(data=titanic, x='pclass', y='age', ax=a1)
sns.boxenplot(data=titanic, x='pclass', y='age', hue='sex', ax=a2)

# lineplotが作った凡例の材料を取得
handles, labels = a2.get_legend_handles_labels()
# 上書きする際に位置を指定
a2.legend(handles, labels, loc='upper left', bbox_to_anchor=(1, 1), frameon=False)
```

```python
sns.catplot(data=titanic, x='pclass', y='age', kind='boxen', hue='survived', col='sex', legend_out=False).fig.axes[0]._loc = 5
```
