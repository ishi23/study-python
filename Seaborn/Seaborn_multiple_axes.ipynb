# multi-plot: 並べる
- 公式　https://seaborn.pydata.org/tutorial/axis_grids.html
- figure-levelの書き方とaxes-levelの書き方があ利、その裏ではここにあるクラスが使われている
- 実際のところ、各関数で並べられるので余程自由度が必要な時以外使わない気がする。
- 飛び道具として、PairGrid()に関してはよく使うと思われる

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
tips.head()
```

## FaceGrid class
- figure-level function
- row, col, hueの三次元まで指定可能で行列化
- map()メソッド引数にaxes-levelの関数を入れてグラフを描写する

```python
g = sns.FacetGrid(data=tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.1)
```

```python
sns.pairplot(iris, hue='species')
```

```python
g = sns.PairGrid(iris, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=0, fill=True)
```
