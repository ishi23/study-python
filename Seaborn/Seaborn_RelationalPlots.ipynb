# relational plot系
- numeric vs numaricの場合、scatter plot系とline plot系がある。
- 上記二種はsns.scatterplot()とsns.lineplot()に相当し、どちらもsns.relplot()で描ける。
- regplot() --regression:回帰-- scatter plotに線形回帰を描写。ax=で並べるタイプ
- lmplot() --linear model:線形モデル-- regplot()のFaceGridで並べるタイプ
- sns.lineplot()公式 https://seaborn.pydata.org/generated/seaborn.lineplot.html
- sns.scatterplot()公式　https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot
- sns.relplot()公式　https://seaborn.pydata.org/generated/seaborn.relplot.html
- sns.regplot()公式　https://seaborn.pydata.org/generated/seaborn.regplot.html#seaborn.regplot
- sns.regplot()公式　https://seaborn.pydata.org/generated/seaborn.regplot.html#seaborn.regplot
- sns.lmplot()公式　https://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

## `sns.lineplot()`折れ線グラフ
- デフォルトでは95%信頼区間を表示する。
- hueで線が別れる。
- その他色々と融通は効く

```python
f, (a1, a2) = plt.subplots(1, 2, figsize=(10,5))
sns.lineplot(data=iris, x='petal_length', y='petal_width', ax=a1)
sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species', ax=a2)
```

## `sns.scatterplot()`散布図
- hueで色、styleでプロット形、sizeでサイズ。
- hueとsizeはnumericでもOK

```python
f, (a1, a2) = plt.subplots(1, 2, figsize=(10,5))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', ax=a1)
sns.scatterplot(data=iris, x='petal_length', y='petal_width'\
                , hue='species', style='species', size='sepal_length', ax=a2)
```

## `sns.relplot()`折れ線と散布図
- lineplot()とscatterplot()と同じことができる。kindで選択。
- col引数で並べられる。categorialのcatplot()に相当

```python
sns.relplot(data=iris, x='petal_length', y='petal_width', kind='line', col='species')
```

```python
sns.relplot(data=iris, x='petal_length', y='petal_width'\
                , kind='scatter', col='species')
```

## `sns.pairplot(df)`散布図行列で全体像を描く。
- デフォルトではnumerialデータを***全て***の組み合わせでプロット
- 各データのヒストグラムも生成
- hueでカテゴリカルデータを選択すると更にいい感じ

```python
sns.pairplot(iris, hue='species')
```

## `sns.jointplot(x, y, data)` 散布図＋ヒストグラムで二変数を可視化

```python
sns.jointplot(data=iris, x='petal_length', y='petal_width', hue='species')
```

## `sns.regplot()` 回帰プロット regression plot
- scatter plotに回帰線を描写
- axで並べるタイプ

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
sns.regplot(data=iris, x='petal_length', y='petal_width', ax=ax1)
sns.regplot(data=iris, x='petal_length', y='petal_width', order=3, ax=ax2)
```

## `sns.lmplot()` 線形モデル linear model
- scatter plotに回帰線を描写
- FaceGridで並べるタイプ

```python
sns.lmplot(data=iris, x='petal_length', y='petal_width', col='species')
```
