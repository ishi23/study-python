## オブジェクト指向で描く

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4,figsize=(20,10))
```

## **plot: 折れ線グラフ**

```python
x = np.linspace(0.1, 5, 10)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.log(x)
y5 = np.log10(x)
y6 = np.log2(x)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(20,10))
axes[0,0].plot(x, y1)
axes[0,1].plot(x, y2)
axes[0,2].plot(x, y3)
axes[1,0].plot(x, y4)
axes[1,1].plot(x, y5)
axes[1,2].plot(x, y6)
axes[2,0].plot(x, y1)
axes[2,0].plot(x, y2)
axes[2,0].plot(x, y3)
axes[2,1].plot(x, y4)
axes[2,1].plot(x, y5)
axes[2,1].plot(x, y6)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20,10))
axes[0].plot(x, y1, label='liner plot')
axes[0].set_xlabel('linspence5', fontsize=20)
axes[0].set_ylabel('y=x', fontsize=20)
axes[0].set_title('y = x', fontsize=30)
axes[0].set_xticks(np.linspace(0,5,11))
axes[0].set_yticks(np.linspace(0,5,6))
axes[0].legend(loc=2, fontsize=15)
```

## for文で回す際は***axes.ravel()***

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
one_dimension_axes = axes.ravel()
x = np.linspace(0, 1, 100)
for i, ax in enumerate(one_dimension_axes):
    ax.plot(x, x ** (i+1))
    ax.set_title('x'+str(i))

plt.show()
```
